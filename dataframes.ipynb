```python
import pandas as pd
import numpy as np

from pydataset import data


np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)
```

### What is the the weight of the fastest animal?

```python
Mammals.sort_values(by='speed', ascending = False).head(1)
```

### What is the overal percentage of specials?

```python
#Calculating the overall percentage of specials by finding all specials and dividing by total number of mammals

overall_pct_specials = (len(Mammals[Mammals.specials == True]) / len(Mammals)) * 100

overall_pct_specials
```

### How many animals are hoppers that are above the median speed? What percentage is this?

```python
#Finding the median speed of all mammals
med_speed = Mammals.speed.median()
med_speed

is_hopper = Mammals[(Mammals.hoppers == True)]

is_above_med =  Mammals[(Mammals.speed > med_speed)]


#Finding the number of hoppers whose speed is above the median speed of all mammals
above_med_hoppers = Mammals[(Mammals.hoppers == True) & (Mammals.speed > med_speed)]

above_med_hoppers


print("Number of hoppers: ", len(is_hopper))

print("Number of mammals that are above the median speed: ", len(is_above_med))

print("Number of hoppers that are above the median speed: ", len(above_med_hoppers))
```

```python
#Calculating the percentage of hoppers above the median speed
pct_above_med = (len(above_med_hoppers) / len(Mammals)) * 100 

pct_above_med
```
