### Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
from env import host, user, password
```

```python
def get_db_url(host, user, password, database):
        
    url = f'mysql+pymysql://{user}:{password}@{host}/{database}'
    
    return url
```

```python
orders = pd.read_sql('SELECT * FROM orders', get_db_url(host,user, password, 'chipotle'))
```

```python
orders.head()
```

```python
top_four = orders.item_name.value_counts().head(4)
top_four
```

```python
# df[column_name] = df.index

#if you need to make a column out of the index quickly, you can do
#df[column_name] = df.index

# df = df.reset_index()
```

```python
top_four["item"] = top_four.index
top_four["item"]
```

```python
orders['item_price'] = orders.item_price.str.replace('$', '', regex=False).astype('float')
```

```python
revenue = orders[orders.item_name.isin(top_four.index)].groupby('item_name').item_price.sum()
revenue = revenue.sort_values(ascending = False).reset_index()
revenue
```

```python
sns.relplot(x='item_name', y = 'item_price', data = revenue)
plt.title('Revenue for Top 4 Most Popular Items')
plt.ylabel("Revenue") 
plt.xlabel("Item Name")
```

```python

sns.barplot(y = 'item_name', x = 'item_price', data = revenue)
plt.title('Most popular items and revenue')

plt.ylabel("") # to get remove the column_name
plt.xlabel("Revenue")
```

### Load the sleepstudy data and read it's documentation. Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant line showing the average change in reaction time.

```python
from pydataset import data

study = data("sleepstudy")

study.head()
```

```python
sns.pairplot(study)
```

```python
study['Subject'] = 'Subject_' + study.Subject.astype(str)
```

```python
plt.figure(figsize=(16, 9))

sns.lineplot(data = study, x = 'Days', y = 'Reaction', hue = 'Subject')
sns.lineplot(data = study, x = 'Days', y = 'Reaction', color = 'navy', estimator = 'mean',linewidth = 4,label='Average')
plt.show()
```
